```python
df2['State_transaction'] = df2['Order Date'].astype(str)+'_'+df2['State'].astype(str)

df2[:5]
```

```python
df3=pd.crosstab(df2['State_transaction'],df2['Product Sub Category'])
```

```python
def encode(item_freq):
    res = 0
    if item_freq > 0:
        res = 1
    return res
    
basket_input = df3.applymap(encode)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
model= apriori(basket_input, min_support=0.001, use_colnames=True)

rules = association_rules(model, metric="lift")

rules.head()
```

```python
rules.sort_values(["support", "confidence","lift"],axis = 0, ascending = False)[:10]
```
